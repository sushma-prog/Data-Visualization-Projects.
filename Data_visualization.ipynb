# Seaborn Basics

## 1. Scatter Plot

```python
!pip install seaborn
```

**data=df: The data source is the DataFrame df, which contains the data to be plotted.**

**x="age": This specifies that the x-axis will represent the "age" column from the DataFrame.**

**y="salary": This specifies that the y-axis will represent the "salary" column from the DataFrame.**

**hue="department": This will color the points based on the "department" column. Different departments will be represented by different colors.**

**style="department": This assigns different marker styles (e.g., circles, squares) to each department, helping to differentiate between departments visually.**

**s=100: This controls the size of the scatter plot markers. A value of 100 makes the points larger.**

```python
import seaborn as sns

# Scatter plot for Age vs. Salary
sns.scatterplot(data=df, x="age", y="salary", hue="department", style="department", s=100)
plt.title("Scatter Plot of age vs salary")
plt.show()
```

## 2. Box Plot

```python
# Box plot of salary by department
sns.boxplot(data=df, x="department", y="salary")
plt.title("Box plot of salary by department")
plt.show()
```
